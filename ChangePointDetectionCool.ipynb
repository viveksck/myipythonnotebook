# Code setup

```python
import pandas as pd
import numpy as np
import pylab as pl
from os import path
```

```python
def load_timeseries(source_timeseries, dest_timeseries):
  sourcedf = pd.read_csv(source_timeseries)
  destdf = pd.read_csv(dest_timeseries)
  return sourcedf, destdf
```

```python
def get_word_list(word_file, num_words =-1):
  """ Returns a list of words for which nearest neighbor computation needs 
      to be done.
  """
  words_list = open(word_file, 'r').read().split('\n')
  if words_list[-1] == '': 
    words_list = words_list[:-1]
  if num_words != -1: 
    return words_list[:num_words]
  else:
    return words_list
```

# Experiment on Randomization

```python
def get_ci(theta_star, theta_hat, nullvalue, blockratio =1.0):
    b_star = np.sort(theta_star[~np.isnan(theta_star)]) # get rid of nans while we sort
    se = np.std(b_star) * np.sqrt(blockratio)
    ci = [b_star[int(len(b_star)*.025)], b_star[int(len(b_star)*.975)]] # bootstrap 95% CI based on empirical percentiles
    # If Null value is present in the CI then we fail p-value test
    if nullvalue > ci[0]:
      return False
    else:
      # We pass the p-value test
      return True
```

```python
def plot_bootstrap(theta_star,  theta_hat,  blockratio=1.0,  bins=10,  title="bootstrapped values"):
    """ 
    given a vector of bootstrap realizations of a statistic, 
    show a histogram with normal and bootstrap confidence intervals. 
    """
    figure()
    scatter(arange(len(theta_star)), theta_star)
    plt.title(title)
    figure()
    b_star = np.sort(theta_star[~np.isnan(theta_star)]) # get rid of nans while we sort
    se = np.std(b_star) * np.sqrt(blockratio)
    ci = [b_star[int(len(b_star)*.025)], b_star[int(len(b_star)*.975)]] # bootstrap 95% CI based on empirical percentiles
    #print "theta_hat = %.4f, E(theta_star)=%.4f, se(theta_hat) = %.4f, 95%% CI = [%.4f, %.4f]" % (
    #        theta_hat, mean(b_star), se, ci[0], ci[1])
    hist(b_star, bins=bins)
    plt.title("counts for %s" % title)
    xlabel("value")
    ylabel("count")
    axvline(x=theta_hat, linestyle='-', color='k')
    axvline(x=mean(b_star), linestyle='-', color='r')
    axvline(x=ci[0], linestyle='--', color='r')
    axvline(x=ci[1], linestyle='--', color='r')    
```

```python
def bootstrap_ts(y, true_val, func, null_value, B = 100, b=3, plot=False):
  beta_star = np.empty(B)
  z = y
  z_star = np.empty(len(z))
  for boot_i in range(B):
      for block_i, start in enumerate(np.random.randint(len(z) - b + 1, size=len(z)/b)):
          z_star[block_i*b:(block_i+1)*b] = z[start:start+b]
      beta_star[boot_i] = func(z_star)
  if get_ci(beta_star, true_val,  null_value, blockratio=float(b)/len(y)):
    if plot:
      plot_bootstrap(beta_star, true_val, blockratio=float(b)/len(y), title="%d moving-block bootstraps, b=%d" % (B, b))
    return True
  else:
    return False
```

```python
def last_time_point(series):
  return series[-1]
```

```python
def slope(series):
  import statsmodels.api as sm
  model = sm.OLS(series, np.arange(0, len(series)))
  results = model.fit()
  return results.params[0]
```

```python
def ar_slope(series):
  import statsmodels.api as sm
  model =  sm.tsa.AR(series)
  results = model.fit()
  return results
```

```python
def get_num_ngrams(w, year, prefix='eng-fiction'):
  input_file = path.join('../mit_lm/topngrams/' + prefix + '/' + year, w + '.txt')
  with open(input_file) as foo:
    lines = foo.readlines()
    if lines:
      return len(lines), lines[0].strip()
    else:
      return 0, None
```

```python
import numpy
def smooth(x,window_len=11,window='hanning'):
        if x.ndim != 1:
                raise ValueError, "smooth only accepts 1 dimension arrays."
        if x.size < window_len:
                raise ValueError, "Input vector needs to be bigger than window size."
        if window_len<3:
                return x
        if not window in ['flat', 'hanning', 'hamming', 'bartlett', 'blackman']:
                raise ValueError, "Window is on of 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'"
        s=numpy.r_[2*x[0]-x[window_len-1::-1],x,2*x[-1]-x[-1:-window_len:-1]]
        if window == 'flat': #moving average
                w=numpy.ones(window_len,'d')
        else:  
                w=eval('numpy.'+window+'(window_len)')
        y=numpy.convolve(w/w.sum(),s,mode='same')
        return y[window_len:-window_len+1]
```

```python
def create_changed_dataframe(csrc, filter_ngrams=False):
    filteredwords=[]
    new_words = []
    for w in csrc.word:
        num_ngrams_src, ngram_src = get_num_ngrams(w, str(csrc.columns[2])) 
        num_ngrams_dest, ngram_dest = get_num_ngrams(w, str(csrc.columns[-1]))
        if num_ngrams_src  == 0:
          new_words.append([w])
        if  filter_ngrams and num_ngrams_src> 90:
          filteredwords.append([w, ngram_src, ngram_dest])
        if  filter_ngrams == False and num_ngrams_src > 0:
          filteredwords.append([w, ngram_src, ngram_dest])
        header = ['word', 'past', 'now']
    cdf = pd.DataFrame.from_records(filteredwords, columns=header)
    newwords_df = pd.DataFrame.from_records(new_words, columns=['word'])
    return cdf, newwords_df
```
