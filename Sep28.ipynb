NER Report
=================================

This documents the results of a grid search performed using neural networks. The following were the parameters and the values that were tried out:

* **Passes** (stochastic gradient): 1, 4, 10
* **Context size**:                 1 2 4
* **Sampling ratio**:               0.001, 0.01, 0.1, 1.
* **Hidden nodes**:                 32, 64, 128, 300
* **Loss functions**: logistic, hinge
* **Learning rate**: Adagrad


Total number of models: **288**

```python
import numpy as np
import pylab as pl
import scipy as scipy
pl.rcParams['figure.figsize'] = 12, 8
```

```python
def LoadFile(filename):
    X = np.genfromtxt(filename, skiprows=1)
    return X

def SelectData(X, col, value):
    return X[X[:, col] == value]

def GetMaxCol(X, colno):
    return np.max(X[:, colno])
```

```python
data_file = './final_2.csv'
```

```python
X = LoadFile(data_file)
sample_ratio_col = 7
context_size_col = 1
f1_score_training_col = 8
f1_score_dev_col = 9
nn_col = 4
f1_score_train_oov_col = 10
f1_score_dev_oov_col = 11
```

<h1> Experiment 1 (effect of context sizes)</h1>
In this experiment I plot the effect of context size assuming a fixed sample of training data(in this case the full training data).<br>
We plot the maximum F1-score achieved for different context sizes

```python
# Plot 1: Plot the effect of context size assuming fixed sample of training data ,
# in this case full training data.

# Select all those rows where sample ratio is 1.0
data_pl_1 = SelectData(X, sample_ratio_col, 1.0)

#Select all those rows where context size is 1,2,4
data_pl_1_ctx1 = SelectData(data_pl_1, context_size_col, 1)
data_pl_1_ctx2 = SelectData(data_pl_1, context_size_col, 2)
data_pl_1_ctx4 = SelectData(data_pl_1, context_size_col, 4)

x_vals = [1,2,4]
map_data =  { 
               1: data_pl_1_ctx1, 
               2: data_pl_1_ctx2,
               4: data_pl_1_ctx4
            }

y_f1_training = []
y_f1_dev = []
y_f1_dev_oov = []
y_f1_train_oov = []
for (ctx_size, source) in sorted(map_data.items()):
    f1_score_training = GetMaxCol(source, f1_score_training_col)
    y_f1_training.append(f1_score_training)
    
    f1_score_dev = GetMaxCol(source, f1_score_dev_col)
    y_f1_dev.append(f1_score_dev)

    f1_score_train_oov = GetMaxCol(source, f1_score_train_oov_col)
    y_f1_train_oov.append(f1_score_train_oov)

    f1_score_dev_oov = GetMaxCol(source, f1_score_dev_oov_col)
    y_f1_dev_oov.append(f1_score_dev_oov)
    


pl.title('F1 score versus context size (sample_ratio: 1.0)')
pl.xlabel('Context size')
pl.ylabel('F1 score')
pl.plot(x_vals, y_f1_training, 'ro-', label='(tr)')
pl.plot(x_vals, y_f1_dev, 'bo-', label='(dev)')
pl.plot(x_vals, y_f1_train_oov, 'yo-' , label ='(tr) OOV')
pl.plot(x_vals, y_f1_dev_oov, 'go-', label='(dev) OOV')
pl.grid()
bla = pl.legend(loc='best')
```

Observations
============

We used F1 reported by conlleval.pl

The plot shows the maximum F1 score obtained for different context sizes.


We note that assuming we get all **OOV words** right(green curve) we see **~3% improvement**.

Context sizes of 2 and 4 look promising. Does not appear much larger context sizes help.

Experiment 2:  (Effect of number of training examples)
=======================================================

```python
# Plot 2: Plot the effect of training data on F1-score. Note this is just 1 run. Variance in
# multiple runs needs to be taken into account

data_pl_2 = X

#Select all those rows where context size is 1,2,4
data_pl_2_s1 = SelectData(data_pl_2, sample_ratio_col, 0.001)
data_pl_2_s2 = SelectData(data_pl_2, sample_ratio_col, 0.01)
data_pl_2_s3 = SelectData(data_pl_2, sample_ratio_col, 0.1)
data_pl_2_s4 = SelectData(data_pl_2, sample_ratio_col, 1.0)

x_vals = [0.001, 0.01, 0.1, 1.0]
map_data =  { 
               0.001: data_pl_2_s1, 
               0.01:  data_pl_2_s2,
               0.1:   data_pl_2_s3,
               1.0:   data_pl_2_s4,
            }

y_f1_training = []
y_f1_dev = []
y_f1_dev_oov = []
y_f1_train_oov = []
for (ratio, source) in sorted(map_data.items()):
    f1_score_training = GetMaxCol(source, f1_score_training_col)
    y_f1_training.append(f1_score_training)
    
    f1_score_dev = GetMaxCol(source, f1_score_dev_col)
    y_f1_dev.append(f1_score_dev)
    
    f1_score_train_oov = GetMaxCol(source, f1_score_train_oov_col)
    y_f1_train_oov.append(f1_score_train_oov)

    
    f1_score_dev_oov = GetMaxCol(source, f1_score_dev_oov_col)
    y_f1_dev_oov.append(f1_score_dev_oov)

pl.title('F1 score versus sample ratio')
pl.xlabel('Sample size')
pl.ylabel('F1 score')
pl.xscale('log')
pl.plot(x_vals, y_f1_training, 'ro-', label='(tr)')
pl.plot(x_vals, y_f1_dev, 'bo-', label='(dev)')
pl.plot(x_vals, y_f1_train_oov, 'yo-' , label ='(tr) OOV')
pl.plot(x_vals, y_f1_dev_oov, 'go-', label='(dev) OOV')
pl.grid()
legend = pl.legend(loc='best')
```

<h2>Observation</h2>
More training data the better as expected.<br>
10% of the training data gets us close to 76.

<h1> Experiment 3: Effect of the number of hidden nodes of the Neural network </h1>

```python
# Plot 3: Plot the effect of number of hidden nodes of NN on F1-score.

data_pl_3 = X

#Select all those rows where context size is 1,2,4
data_pl_3_32 = SelectData(data_pl_3, nn_col, 32)
data_pl_3_64 = SelectData(data_pl_3, nn_col, 64)
data_pl_3_128 = SelectData(data_pl_3, nn_col, 128)
data_pl_3_300 = SelectData(data_pl_3, nn_col, 300)

x_vals = [32, 64, 128, 300]
map_data =  { 
               32:   data_pl_3_32, 
               64:   data_pl_3_64,
               128:  data_pl_3_128,
               300:  data_pl_3_300,
            }

y_f1_training = []
y_f1_dev = []
y_f1_dev_oov = []
y_f1_train_oov = []
for (ratio, source) in sorted(map_data.items()):
    f1_score_training = GetMaxCol(source, f1_score_training_col)
    y_f1_training.append(f1_score_training)
    
    f1_score_dev = GetMaxCol(source, f1_score_dev_col)
    y_f1_dev.append(f1_score_dev)
    
    f1_score_train_oov = GetMaxCol(source, f1_score_train_oov_col)
    y_f1_train_oov.append(f1_score_train_oov)

    f1_score_dev_oov = GetMaxCol(source, f1_score_dev_oov_col)
    y_f1_dev_oov.append(f1_score_dev_oov)

    

pl.title('F1 score versus NN hidden nodes')
pl.xlabel('Number of hidden nodes')
pl.ylabel('F1 score')
pl.xscale('log', basex=2)
pl.plot(x_vals, y_f1_training, 'ro-', label='(tr)')
pl.plot(x_vals, y_f1_dev, 'bo-', label='(dev)')
pl.plot(x_vals, y_f1_train_oov, 'yo-' , label ='(tr) OOV')
pl.plot(x_vals, y_f1_dev_oov, 'go-', label='(dev) OOV')
pl.grid()
legend = pl.legend(loc='best')
```
