```python
import numpy as np
import pylab as pl
import scipy as scipy
```

```python
def LoadFile(filename):
    X = np.genfromtxt(filename, skiprows=1)
    return X

def SelectData(X, col, value):
    return X[X[:, col] == value]

def GetMaxCol(X, colno):
    return np.max(X[:, colno])
```

```python
data_file = './final28.csv'
```

```python
X = LoadFile(data_file)
sample_ratio_col = 7
context_size_col = 1
f1_score_training_col = 8
f1_score_dev_col = 9
nn_col = 4
```

```python
# Plot 1: Plot the effect of context size assuming fixed sample of training data ,
# in this case full training data.

# Select all those rows where sample ratio is 1.0
data_pl_1 = SelectData(X, sample_ratio_col, 1.0)

#Select all those rows where context size is 1,2,4
data_pl_1_ctx1 = SelectData(data_pl_1, context_size_col, 1)
data_pl_1_ctx2 = SelectData(data_pl_1, context_size_col, 2)
data_pl_1_ctx4 = SelectData(data_pl_1, context_size_col, 4)

x_vals = [1,2,4]
map_data =  { 
               1: data_pl_1_ctx1, 
               2: data_pl_1_ctx2,
               4: data_pl_1_ctx4
            }

y_f1_training = []
y_f1_dev = []
for (ctx_size, source) in sorted(map_data.items()):
    f1_score_training = GetMaxCol(source, f1_score_training_col)
    y_f1_training.append(f1_score_training)
    
    f1_score_dev = GetMaxCol(source, f1_score_dev_col)
    y_f1_dev.append(f1_score_dev)

pl.title('F1 score versus context size (sample_ratio: 1.0)')
pl.xlabel('Context size')
pl.ylabel('F1 score')
pl.plot(x_vals, y_f1_training, 'ro-', label='F1 (training)')
pl.plot(x_vals, y_f1_dev, 'bo-', label='F1 (dev)')
pl.legend(loc='best')
```

```python
# Plot 2: Plot the effect of training data on F1-score. Note this is just 1 run. Variance in
# multiple runs needs to be taken into account

data_pl_2 = X

#Select all those rows where context size is 1,2,4
data_pl_2_s1 = SelectData(data_pl_2, sample_ratio_col, 0.001)
data_pl_2_s2 = SelectData(data_pl_2, sample_ratio_col, 0.01)
data_pl_2_s3 = SelectData(data_pl_2, sample_ratio_col, 0.1)
data_pl_2_s4 = SelectData(data_pl_2, sample_ratio_col, 1.0)

x_vals = [0.001, 0.01, 0.1, 1.0]
map_data =  { 
               0.001: data_pl_2_s1, 
               0.01:  data_pl_2_s2,
               0.1:   data_pl_2_s3,
               1.0:   data_pl_2_s4,
            }

y_f1_training = []
y_f1_dev = []
for (ratio, source) in sorted(map_data.items()):
    f1_score_training = GetMaxCol(source, f1_score_training_col)
    y_f1_training.append(f1_score_training)
    
    f1_score_dev = GetMaxCol(source, f1_score_dev_col)
    y_f1_dev.append(f1_score_dev)

pl.title('F1 score versus sample ratio')
pl.xlabel('Sample size')
pl.ylabel('F1 score')
pl.xscale('log')
pl.plot(x_vals, y_f1_training, 'ro-', label='F1 (training)')
pl.plot(x_vals, y_f1_dev, 'bo-', label='F1 (dev)')
pl.legend(loc='best')
```

```python
# Plot 3: Plot the effect of number of hidden nodes of NN on F1-score.

data_pl_3 = X

#Select all those rows where context size is 1,2,4
data_pl_3_32 = SelectData(data_pl_3, nn_col, 32)
data_pl_3_64 = SelectData(data_pl_3, nn_col, 64)
data_pl_3_128 = SelectData(data_pl_3, nn_col, 128)
data_pl_3_300 = SelectData(data_pl_3, nn_col, 300)

x_vals = [32, 64, 128, 300]
map_data =  { 
               32:   data_pl_3_32, 
               64:   data_pl_3_64,
               128:  data_pl_3_128,
               300:  data_pl_3_300,
            }

y_f1_training = []
y_f1_dev = []
for (ratio, source) in sorted(map_data.items()):
    f1_score_training = GetMaxCol(source, f1_score_training_col)
    y_f1_training.append(f1_score_training)
    
    f1_score_dev = GetMaxCol(source, f1_score_dev_col)
    y_f1_dev.append(f1_score_dev)

pl.title('F1 score versus NN hidden nodes')
pl.xlabel('Number of hidden nodes')
pl.ylabel('F1 score')
pl.xscale('log', basex=2)
pl.plot(x_vals, y_f1_training, 'ro-', label='F1 (training)')
pl.plot(x_vals, y_f1_dev, 'bo-', label='F1 (dev)')
pl.legend(loc='best')
```
