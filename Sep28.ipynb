<h1> Brief summary of the grid search </h1>This documents the results of a grid search performed using neural networks. The following were the parameters and the values that were tried out:1. Number of passes (stochastic gradient): 1,4,102. Context size: 1 2 43. Sample proportion of training data: 0.001, 0.01, 0.1, 1. We just randomly sample from the training data4. Number of hidden nodes in the neural network: 32, 64, 128, 3005. Loss functions: 'logistic' , 'hinge'6. Learning rate: Used exact adaptive norm learning rateTotal number of models: <b>288</b>

```python
import numpy as npimport pylab as plimport scipy as scipy
```

```python
def LoadFile(filename):    X = np.genfromtxt(filename, skiprows=1)    return Xdef SelectData(X, col, value):    return X[X[:, col] == value]def GetMaxCol(X, colno):    return np.max(X[:, colno])
```

```python
data_file = './final_2.csv'
```

```python
X = LoadFile(data_file)sample_ratio_col = 7context_size_col = 1f1_score_training_col = 8f1_score_dev_col = 9nn_col = 4f1_score_train_oov_col = 10f1_score_dev_oov_col = 11
```

<h1> Experiment 1 (effect of context sizes)</h1>In this experiment I plot the effect of context size assuming a fixed sample of training data(in this case the full training data).<br>We plot the maximum F1-score achieved for different context sizes

```python
# Plot 1: Plot the effect of context size assuming fixed sample of training data ,# in this case full training data.# Select all those rows where sample ratio is 1.0data_pl_1 = SelectData(X, sample_ratio_col, 1.0)#Select all those rows where context size is 1,2,4data_pl_1_ctx1 = SelectData(data_pl_1, context_size_col, 1)data_pl_1_ctx2 = SelectData(data_pl_1, context_size_col, 2)data_pl_1_ctx4 = SelectData(data_pl_1, context_size_col, 4)x_vals = [1,2,4]map_data =  {                1: data_pl_1_ctx1,                2: data_pl_1_ctx2,               4: data_pl_1_ctx4            }y_f1_training = []y_f1_dev = []y_f1_dev_oov = []y_f1_train_oov = []for (ctx_size, source) in sorted(map_data.items()):    f1_score_training = GetMaxCol(source, f1_score_training_col)    y_f1_training.append(f1_score_training)        f1_score_dev = GetMaxCol(source, f1_score_dev_col)    y_f1_dev.append(f1_score_dev)    f1_score_train_oov = GetMaxCol(source, f1_score_train_oov_col)    y_f1_train_oov.append(f1_score_train_oov)    f1_score_dev_oov = GetMaxCol(source, f1_score_dev_oov_col)    y_f1_dev_oov.append(f1_score_dev_oov)    pl.title('F1 score versus context size (sample_ratio: 1.0)')pl.xlabel('Context size')pl.ylabel('F1 score')pl.plot(x_vals, y_f1_training, 'ro-', label='(tr)')pl.plot(x_vals, y_f1_dev, 'bo-', label='(dev)')pl.plot(x_vals, y_f1_train_oov, 'yo-' , label ='(tr) OOV')pl.plot(x_vals, y_f1_dev_oov, 'go-', label='(dev) OOV')pl.grid()pl.legend(loc='best')
```

<h2> Observations </h2>F1 score is the score reported by the standard evaluation script of CONLL. <br>The plot shows the maximum F1 score obtained for different context sizes. <br>We note that assuming we get all OOV words right(green curve) we see ~3% improvement. <br>Context sizes of 2 and 4 look promising. Does not appear much larger context sizes help.
