# Changepoints of words

In this experiment, I seek to identify the approximate change point of words. 
I also bin the changed words by decade to get an idea of which decades have contributed to the most change

```python
words=['peck', 'transmitted', 'acid', 'nursing', 'peer', 'mail', 'scanned', 'gay', 'steering', 'receiver']
words.extend(['bet', 'checking', 'babe', 'date', 'dial', 'recording', 'honey', 'bat', 'tissue', 'zone'])
words.extend(['passenger', 'sex', 'diet', 'screw', 'cell', 'boarding', 'stored', 'dumb', 'thumbs'])
words.extend(['grade','spinning','crush', 'mate', 'ass', 'cock', 'calls', 'wrap', 'binding', 'checked'])
words.extend(['hook', 'answering', 'bolt', 'pepper', 'drugs', 'glove', 'drug', 'spine', 'pregnant'])
```

```python
df = pd.read_csv('./5000wordsWithChangePointsFilter.csv')
```

```python
df[df.word.isin(words)]
```

```python
df[df.changepoint.between(1970,1980)]
```

```python
syears = [1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000]
eyears = [ s + 9 for s in syears]
decade_counts=[]
for sy, ey in zip(syears, eyears):
  decade_counts.append(len(df[df.changepoint.between(sy,  ey)]))
```

```python
bar(syears, decade_counts,)
title('Distribution of words that have changed according to decade')
```
