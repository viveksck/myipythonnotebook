```python
def GetLabelDistribution(data_file):
    cmd = "cut -f 2 " +  data_file + " | sort -n | uniq -c"
    output = subprocess.check_output(cmd, shell=True)
    L = output.strip().split()
    tags = L[2::2]
    counts_ = L[1::2]
    counts = [float(count) for count in counts_]
    print "Distribution of labels:", tags, counts
    probs_ = counts/numpy.sum(counts)
    print "Probabilities of labels:",probs_
    return zip(tags, counts), zip(tags, probs_)
```

```python
def GetSampleDict(raw_training_file, sample_threshold, total_num_examples, non_entity_tag):
    tags_counts, tags_probs = GetLabelDistribution(raw_training_file)
    tags_counts_map = dict(tags_counts)
    tags_probs_map = dict(tags_probs)
    entity_counts = [(tag,count) for tag,count  in tags_counts_map.items() if tag != non_entity_tag]
    total_entities = numpy.sum((zip(*entity_counts)[1]))
    prob_entity_counts = [(tag,(count/total_entities)) for tag, count in entity_counts]
    sample_non_entities = int(sample_threshold * total_num_examples)
    remaining_examples = total_num_examples -sample_non_entities
    sample_entities  = [(tag, int(prob * remaining_examples)) for tag, prob in prob_entity_counts]
    sample_dict = dict(sample_entities)
    sample_dict[non_entity_tag] = sample_non_entities
    return sample_dict

def CreateTempFileForTag(raw_training_file, tag, number_of_lines):
    _, temp_file = tempfile.mkstemp(suffix='.conll', dir = working_dir, text=True)
    os.close(_)
    cmd = 'grep ' + tag + "$ " + raw_training_file + "|" + "shuf -n " + str(number_of_lines) + " -o " + temp_file
    output = subprocess.check_output(cmd,shell=True)
    return temp_file

def SampleTrainingFile(raw_training_file, total_num_examples):
    _, temp_file = tempfile.mkstemp(suffix='.conll', dir = working_dir, text=True)
    os.close(_)
    cmd = "shuf -n " + str(total_num_examples) + " -o " + temp_file + " " + raw_training_file
    print cmd
    output = subprocess.check_output(cmd, shell=True)
    D = {}
    D['full_training'] = temp_file
    return D

def OverSampleTrainingFile(raw_training_file, sample_threshold = 0.0, total_num_examples = 200000, non_entity_tag='O'):
    sample_dict = GetSampleDict(raw_training_file, sample_threshold, total_num_examples, non_entity_tag)
    temp_file_dict = {}
    for tag, sample in sample_dict.items():
        temp_file_dict[tag] = CreateTempFileForTag(raw_training_file, tag, sample)
    
    # Create a temporary file to hold the full training
    _, temp_file = tempfile.mkstemp(suffix='.conll', dir = working_dir, text=True)
    os.close(_)
    # Concatenate all the parts
    catf = ' '.join(temp_file_dict.values())
    cmd = 'cat ' + catf + '| shuf > ' + temp_file
    subprocess.check_output(cmd,shell=True)
    
    #Store it in the dictionary
    temp_file_dict['full_training'] = temp_file
    return temp_file_dict

# Do nothing
def DummyTrainingFileFunc(raw_training_file, dummy_arg):
    D = {}
    D['full_training'] = raw_training_file
    return D
```

```python
def GetFeatures(training_file, embeddings, context_size=2):
    sents, tags = read_ner_conll([training_file])
    examples = get_examples(sents, tags, embeddings, context_size)
    vectors, labels, words = zip(*examples)
    features = [{i:v for i, v in enumerate(vector)} for vector in vectors]
    sentence_lengths = [len(x) for x in sents]
    return features, labels, words, sentence_lengths

    
def TrainModel(training_file, embeddings):
    features, labels, words,_ = GetFeatures(training_file, embeddings)    
    num_classes = len(set(labels))
    print "Number of classes:",num_classes
    id_ = random.randint(100000, 999999)
    #Train a model
    clf = VW_Classifier(nn = 32,
                        loss='hinge', 
                        moniker='vw_sklearn{}'.format(id_),
                        passes=10, 
                        silent=True, 
                        oaa=num_classes,
                        adaptive=True,
                        working_dir=working_dir)
    start = time()
    clf.fit(features, labels)
    print("Done training in {}".format(time()-start))
    return clf

def EvaluateModel(model, evaluation_file,embeddings):
    features, labels, words, lengths = GetFeatures(evaluation_file, embeddings)
    y_pred = model.predict(features).tolist()
    y_pred_ = [ID_TAG[x] for x in y_pred]
    labels_ = [ID_TAG[x] for x in labels]
    pairs = zip(words, labels_, y_pred_)
    boundaries = [0]+[sum(lengths[:i]) for i in range(1, len(lengths)+1)]
    sent_pred = [pairs[x:y] for x,y in zip(boundaries[:-1], boundaries[1:])]
    scores, oov_scores = EvaluateConLLEx(sent_pred, working_dir)
    return scores, oov_scores, sent_pred

def RemoveTempFiles(temp_file_dict):
    for temp_file in temp_file_dict.values():
        os.unlink(temp_file)
    return
    
```

```python
def RunConfiguration(key, lang, parts, training_datasets_dir,
                     TrainingFileFunc, FuncArgs, dev_datasets_dir, 
                     embeddings, delete_temp_files = False):
    training_file = path.join(training_datasets_dir, lang, '{}_{}.conll'.format(lang, parts[0]))
    temp_file_dict = TrainingFileFunc(training_file, *FuncArgs)
    print "Created training file"
    clf = TrainModel(temp_file_dict['full_training'], preloaded_embeddings[lang]['embeddings'][embeddings])
    print "Trained the model"
    evaluation_file = path.join(dev_datasets_dir, lang, '{}_{}.conll'.format(lang, parts[1]))
    scores, oov_scores, sent_pred = EvaluateModel(clf, evaluation_file, preloaded_embeddings[lang]['embeddings'][embeddings])
    if delete_temp_files:
        RemoveTempFiles(temp_file_dict)
    return clf, scores, oov_scores, sent_pred
```

```python
def RunExperiment(configurations, output_file):
    # Dump the input configurations
    output_file_path = path.join(results_dir, output_file)
    with open(output_file_path,'wb') as f:
        pickle.dump(configurations, f)
        result_configurations = {}
        for key in configurations:
            try:
                clf, scores, oov_scores, sent_pred = RunConfiguration(key, *configurations[key])
                result_configurations[key] = [clf, scores, oov_scores, sent_pred]
                #Dump the intermediate result as well
            except:
                print "Exception raised for", key
                pass
        pickle.dump(result_configurations, f)
        #We are all done
    return result_configurations
```

```python
def GetScores(exp1_index, j,  D1, D2, D3):
    from scipy.stats import sem
    en_scores = []
    exp1_scores = D1[exp1_index][1]['mean_f1']
    
    L = [D2[j][1]['mean_f1'], D2[j + 1][1]['mean_f1'], D2[j + 2][1]['mean_f1']]
    exp2_scores = numpy.mean(L)
    exp2_err = sem(L)
    
    M = [D3[j][1]['mean_f1'], D3[j + 1][1]['mean_f1'], D3[j + 2][1]['mean_f1']]
    exp3_scores = numpy.mean(M)
    exp3_err = sem(M)
    
    return [exp1_scores, exp2_scores, exp3_scores], [0.0, exp2_err, exp3_err]
    

#!/usr/bin/env python
# a bar plot with errorbars
import numpy as np
import matplotlib.pyplot as plt
import pickle

exp1_file_path = path.join(results_dir, 'exp1_new_1.pkl') 
exp2_file_path = path.join(results_dir, 'exp2_new_1.pkl')
exp3_file_path = path.join(results_dir, 'exp3_new_1.pkl')

exp1_f = open(exp1_file_path,'rb')
exp2_f = open(exp2_file_path,'rb')
exp3_f = open(exp3_file_path,'rb')

D1 = pickle.load(exp1_f)
D1_1 = pickle.load(exp1_f)

D2 = pickle.load(exp2_f)
D2_1 = pickle.load(exp2_f)

D3 = pickle.load(exp3_f)
D3_1 = pickle.load(exp3_f)

en_scores, en_err =  GetScores(1, 1, D1_1, D2_1, D3_1)
es_scores, es_err =  GetScores(2, 4, D1_1, D2_1, D3_1)
nl_scores, nl_err =  GetScores(3, 7, D1_1, D2_1, D3_1)
```

```python
N = 3
ind = np.arange(N)  # the x locations for the groups
width = 0.15       # the width of the bars
fig, ax = plt.subplots()
rects1 = ax.bar(ind, en_scores, width, color='r', yerr = en_err)
rects2 = ax.bar(ind+ 1 * width, es_scores, width, color='y', yerr = es_err)
rects3 = ax.bar(ind+ 2 * width, nl_scores, width, color='b', yerr = nl_err)

# add some
ax.set_ylabel('Scores')
ax.set_title('Scores by language')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('ConLL', 'Random', 'Oversampled') )

ax.legend( (rects1[0], rects2[0], rects3[0]), ('en', 'es', 'nl') )

plt.show()
```

# Failure Analysis of results on Wikipedia

We would like to analyze which examples we got wrong and understand if there is a systemic error

```python
import pickle
```

```python
def LoadResultData(results_file):
    results_file_handle = open(results_file,'rb')
    config = pickle.load(results_file_handle)
    results = pickle.load(results_file_handle)
    return results

def GetWrongPredictions(results_pred, true_tag_filter):
    wrong_predictions = []
    for sentence in results_pred:
        for (word, true_tag, pred_tag) in sentence:
            if true_tag != pred_tag and true_tag_filter == true_tag:
                wrong_predictions.append((word, true_tag, pred_tag))
    return wrong_predictions
```
