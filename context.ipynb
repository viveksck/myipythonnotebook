```python
import numpy
import matplotlib
import pylab as pl
```

```python
def ExtractContextandScores(input_file, field_context_size, field_f_score):
    data = numpy.genfromtxt(input_file, skiprows=1)
    data_1 = data[data[:, field_context_size] == 1]
    data_2 = data[data[:, field_context_size] == 2]
    data_4 = data[data[:, field_context_size] == 4]
    
    f1score_data_1 = numpy.max(data_1[:, field_f_score])
    f1score_data_2 = numpy.max(data_2[:, field_f_score])
    f1score_data_4 = numpy.max(data_4[:, field_f_score])
    
    return [1, 2, 4] , [f1score_data_1, f1score_data_2, f1score_data_4]
```

```python
## Exp 2: use NN 16,32,64,128 nodes NN (without L1 and L2 regularization) 
exp2_file_galileo = "/home/vivek/ner_plots/exp2_galileo.csv"
field_context_size = 1
field_f1_score = 8
x,y = ExtractContextandScores(exp2_file_galileo, field_context_size, field_f1_score)
pl.title("Average F1-Score versus context size (NN : 16,32, 64, 128)")
pl.xlabel("Context size")
pl.ylabel("Average F1-Score")
pl.plot(x,y)
```

```python
## Exp 2: use logistic and hinge losses (with L1 and L2 regularization) 
exp1_file_einstein = "/home/vivek/ner_plots/exp1_einstein.csv"
field_context_size = 1
field_f1_score = 10
x,y = ExtractContextandScores(exp1_file_einstein, field_context_size, field_f1_score)
pl.title("Average F1-Score versus context size (Logistic and Hinge loss)")
pl.xlabel("Context size")
pl.ylabel("Average F1-Score")
pl.plot(x,y)
```

```python
## Exp 3: use logistic and hinge losses (with L1 and L2 regularization) with weighted examples
## based on their frequency
exp2_file_einstein = "/home/vivek/ner_plots/exp2_einstein.csv"
field_context_size = 1
field_f1_score = 10
x,y = ExtractContextandScores(exp2_file_einstein, field_context_size, field_f1_score)
pl.title("Average F1-Score versus context size (Logistic and Hinge loss)")
pl.xlabel("Context size")
pl.ylabel("Average F1-Score")
pl.plot(x,y)
```
