```python
import numpy
import matplotlib
import pylab as pl
from matplotlib.font_manager import FontProperties
```

```python
def ExtractContextandScores(input_file, field_context_size, field_f_score):
    data = numpy.genfromtxt(input_file, skiprows=1)
    data_1 = data[data[:, field_context_size] == 1]
    data_2 = data[data[:, field_context_size] == 2]
    data_4 = data[data[:, field_context_size] == 4]
    
    f1score_data_1 = numpy.max(data_1[:, field_f_score])
    f1score_data_2 = numpy.max(data_2[:, field_f_score])
    f1score_data_4 = numpy.max(data_4[:, field_f_score])
    
    return [1, 2, 4] , [f1score_data_1, f1score_data_2, f1score_data_4]
```

```python
## Exp 2: use NN 16,32,64,128 nodes NN (without L1 and L2 regularization) 
exp2_file_galileo = "/home/vivek/ner_plots/exp2_galileo.csv"
exp1_file_einstein = "/home/vivek/ner_plots/exp1_einstein.csv"
exp2_file_einstein = "/home/vivek/ner_plots/exp2_einstein.csv"
x_exp2g,y_exp2g = ExtractContextandScores(exp2_file_galileo, 1, 8)
x_exp1e,y_exp1e = ExtractContextandScores(exp1_file_einstein, 1, 10)
x_exp2e,y_exp2e = ExtractContextandScores(exp2_file_einstein, 1, 10)
fig,ax = pl.subplots()
ax.plot(x_exp2g, y_exp2g, 'ks', label='NN')
ax.plot(x_exp1e, y_exp1e, 'rs', label='Hinge and logistic loss')
ax.plot(x_exp2e, y_exp2e, 'bs', label='Hinge and logistic loss(weighted)')
pl.xlim((0,6))
pl.legend(loc=7)
pl.title("Average F1-Score versus context size")
pl.xlabel("Context size")
pl.ylabel("F1-Score")
pl.show()
```
