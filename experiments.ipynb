# Data set

***The CIFAR-10 data is very widely used and is a standard data set used for image recognition. It has a training set of 60000 32X32 color images and 10000 test images. The images are from 10 classes like airplane, automobile, bird, cat, deer etc. You can read more here: http://www.cs.toronto.edu/~kriz/cifar.html. This has been used in several deep learning papers as well***

**To investigate the simplest possible case, I just adapted Hintons code(as is) for training a DBN which worked for MNIST data set on the above CIFAR data set. Note that this still uses the original conjugate gradient descent optimization. This was to get a baseline. Training the deep belief network to classify CIFAR-10 for 100 epochs resulted in the following:
a. I noticed that the training error had not gone down to a sufficiently low value. For example, after 100 epochs the training error was 27247 which is pretty high. The corresponding test error was 5762.**

**My investigations led me to conclude that Deep Belief Networks used for MNIST does not perform well on more complex data sets which contain natural images. Refer this link: http://deeplearning.net/tutorial/lenet.html .As I understand, convolutional neural nets are more powerful in the sense that hey are able to extract better features(for example edge detectors etc) from images and better exploit local correlation present.**

**Thus I decided to look at using convolutional nets (CNNs) for this task and am looking at incorporating our algorithm (which adds noise to gradients and performs max norm regularization when updating weights) when training CNN's.**

** I could find a implementation of CNN's which looked reasonablle to me (and something I felt could fit well for adaptation): https://code.google.com/p/cuda-convnet/**

** I made changes to do the above toolkit to adapt it to use our weight update rule **

** I used a already available configuration of a CNN with 3 convolutional layers(conv1, conv2, conv3), and 2 Fully Connected layers (of size 64 and 10 respectively: fc64 and fc10) available in the above toolkit**.

** Below I show some preliminary results (with some details below):
   a. To ensure we get to the right convex regime the first 5 epochs (epoch 0 through 4) are standard (no noise or max norm regularization takes place) (the pretraining stage if you call it).
   b. Subsequent epochs add Gaussian noise with specified standard deviations and perform max-norm regularization.
   My initial trials led to a reasonable choice of C = 100.
