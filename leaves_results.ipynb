```python
import numpy
from os import path
import pickle
```

```python
import pandas as pd
```

```python
output_dir= './output/'
```

```python
rows = []
for i in range(0, 50):
    ofilename = path.join(output_dir, 'result' + str(i) + '.pkl')
    f = pickle.load(open(ofilename))
    grid_point = i
    max_norm = f[0]['max_norm']
    noises = str(f[0]['noises'])
    fine_tune_err = f[1].errors_fine_tune_
    test_err = f[7]
    row  = [grid_point, max_norm, noises, fine_tune_err[f[1].epochs - 1], test_err]
    rows.append(row)
```

```python
header = ['GridPoint', 'C', 'Noises', 'Training Err', 'Test Err']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_records(rows, columns = header)
```

# Result of grid search

** The data contains images of 10 different types of leaves (like maple, oak, pear etc). I have ~9K suchh training example. I split the data into 2 parts (randomly): namely for training and test. The test contains 1000 examples. Each image is 32X32(1024 pixels). **

** I used a Deep Belief Network of the following architecture [1024, 1024, 1024, 10] and tried different parameters of noise and max norm regularization (C which constrains the weights to a ball of radius C). Below are the results of the grid search for different parameters. I ran this for 30 epochs. One can see that this converges. See plot below and note that training error is low. **

** The first row : Grid point 0 shows the training and test errors for just plain SGD (no changes from our side). So this can be base line. **

** We note that adding noise seems to help again in general. If one looks at the test_errs in general one finds adding noise helps as long as the regularization is not too aggressive. **

** Note several grid points perform better than base line (0.090) like Grid point 2, 3, 4, 37, 49.**

** Very aggressive regularization (C = [1,2,4] etc) dont perform as well. The high training error indicates that we are in bias mode by going overboard.**

```python
def plot_training_err(base, num):
    bfilename = path.join(output_dir, 'result' + str(base) + '.pkl')
    bf = pickle.load(open(bfilename))
    b_training_err = bf[1].errors_fine_tune_

    ofilename = path.join(output_dir, 'result' + str(num) + '.pkl')
    f = pickle.load(open(ofilename))
    training_err = f[1].errors_fine_tune_
    
    plot(arange(0, len(training_err)), b_training_err, arange(0, len(training_err)), training_err, linewidth = 4, label=['baseline','config'])
    xlabel('Epochs')
    ylabel('Training Error')
    title('Training error')
    legend()
```

```python
%pylab inline
plot_training_err(0, 37)
```

** The above plot shows the training error curve obtaind for baseline (grid point 0) and the best number we get (grid point 37)**
